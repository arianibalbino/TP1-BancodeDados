## A0. π_nome completo σ_idade≥80 (pessoa ⋊⋉ jogos competidor)

```python
query = """
SELECT 
  P.nome_completo
  --JC.idade
FROM 
  pessoa P
INNER JOIN 
  jogos_competidor JC
ON 
  JC.id_pessoa = P.id_pessoa
WHERE
  JC. idade >= 80
GROUP BY
  P.nome_completo
ORDER BY
  JC.idade DESC

"""

df = pd.read_sql_query(query, conn)
df
```

## A1. π_nome completo σ_nome regiao=′Equipe olímpica de refugiados′ (pessoa ⋊⋉ pessoa_regiao ⋊⋉ regiao_con)


```python
query = """
SELECT 
  P.nome_completo
FROM 
  pessoa P
INNER JOIN 
  pessoa_regiao PR
ON 
  PR.id_pessoa = P.id_pessoa
INNER JOIN
  regiao_con RC
ON
  PR.id_regiao = RC.id_regiao
WHERE
  RC.nome_regiao= 'Equipe olímpica de refugiados'
GROUP BY
  P.nome_completo


"""

df = pd.read_sql_query(query, conn)
df
```

## A2. π_ano_jogos,estacao σ_nome cidade=′Londres′ (jogos ⋊⋉ jogos cidade ⋊⋉ cidade)


```python
query = """
SELECT 
  J.ano_jogos,
  J.estacao
  --C.nome_cidade
FROM 
  jogos J
INNER JOIN 
  jogos_cidade JC
ON 
  J.id_jogos = JC.id_jogos
INNER JOIN
  cidade C
ON
  C.id_cidade = JC.id_cidade
WHERE
  C.nome_cidade= 'Londres'



"""

df = pd.read_sql_query(query, conn)
df
```

## A3. π_nome_completo,altura σ_altura≥215 ∧ ano_jogos=2016 (pessoa ⋊⋉ jogos_competidor ⋊⋉ jogos)



```python
query = """
SELECT 
  P.nome_completo,
  P.altura
  --J.ano_jogos
FROM 
  pessoa P
INNER JOIN 
  jogos_competidor Jcon
ON 
  Jcon.id_pessoa = P.id_pessoa
INNER JOIN
  jogos J
ON
  J.id_jogos = Jcon.id_jogos
WHERE
  P.altura >= 215
AND
  J.ano_jogos = 2016

GROUP BY
  P.nome_completo


"""

df = pd.read_sql_query(query, conn)
df
```

## A4. π_nome_completo,altura σ_genero= ′F ′ ∧ altura≥200 (pessoa)

## B0. Liste (1) o nome dos esportes que nao sao disputados em Jogos Olimpicos de Verao. O nome dos esportes deve aparecer apenas uma vez, para cada esporte listado.



```python
query = """
SELECT DISTINCT
  E.nome_esporte
  --J.estacao
FROM 
  esporte E
INNER JOIN
  Evento Ev 
ON
  E.id_esporte = Ev.id_esporte
INNER JOIN
  competidor_evento CE
ON  
  CE.id_evento = Ev.id_evento
INNER JOIN
  jogos_competidor Jcon
ON
  Jcon.id_competidor = CE.id_competidor
INNER JOIN
  jogos J
ON
  J.id_jogos = Jcon.id_jogos
WHERE
  J.estacao <> 'Verão'



"""

df = pd.read_sql_query(query, conn)
df
```

## B2. Liste (1) o codigo e (2) o nome dos cinco CONs com o maior numero de atletas registrados. Informe tambem (3) a quantidade total de atletaspara cada CON.



```python
query = """
SELECT 
  RC.con,
  RC.nome_regiao,
  COUNT (PR.id_pessoa) AS qtd
FROM 
  regiao_con RC
INNER JOIN
  pessoa_regiao PR 
ON
  RC.id_regiao = PR.id_regiao

GROUP BY
  RC.con
ORDER BY
  qtd
DESC

LIMIT 5




"""

df = pd.read_sql_query(query, conn)
df
```

## B3. Liste (1) o nome das cidades que sediaram mais de uma edicao de JogosOlimpicos e (2) seus respectivos paises.

```python
query = """
SELECT DISTINCT
  C.nome_cidade,
  C.nome_pais
FROM 
  cidade C
INNER JOIN
  jogos_cidade JC
ON
  C.id_cidade = JC.id_cidade
 INNER JOIN
  jogos J
ON
  J.id_jogos = JC.id_jogos
GROUP BY
  C.nome_cidade
 HAVING COUNT (J.ano_jogos) > 1

"""

df = pd.read_sql_query(query, conn)
df
```

## B4. Liste (1) o codigo e (2) o nome dos CONs que ganharam pelo menos uma medalha (i.e., Ouro, Prata ou Bronze) na historia dos Jogos Olimpicos.


```python
query = """
 SELECT DISTINCT
  RC.con,
  RC.nome_regiao
  --M.nome_medalha,
  --COUNT (M.id_medalha) AS qtd
FROM
  regiao_Con RC
  INNER JOIN
    pessoa_regiao PR
  ON
    RC.Id_regiao = PR.id_regiao
  INNER JOIN
    pessoa  P
  ON
    P.Id_pessoa = PR.id_pessoa
  INNER JOIN
    jogos_competidor JC
  ON
    JC.id_pessoa = PR.Id_pessoa
  INNER JOIN
    competidor_evento  CE
  ON
    JC.Id_competidor = CE.id_competidor
  INNER JOIN
      medalha M
    ON
      CE.id_medalha = M.Id_medalha
WHERE
  M.nome_medalha != 'NA' 
GROUP BY
  RC.con




"""

df = pd.read_sql_query(query, conn)
df
```

## B5. Liste (1) o nome dos paises e (2) o continente com cidades sede de Jogos Olimpicos de Inverno.



```python
query = """
SELECT DISTINCT
  C.nome_pais,
  C.nome_continente
FROM 
  cidade C
INNER JOIN
  jogos_cidade JC 
ON
  JC.id_cidade = C.id_cidade
INNER JOIN
  jogos J
ON
  J.id_jogos = JC.id_jogos
WHERE
  J.estacao = 'Inverno'
GROUP BY
  C.nome_cidade


"""

df = pd.read_sql_query(query, conn)
df
```
